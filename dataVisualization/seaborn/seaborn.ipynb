#### Heatmaps
graphical reresentation of individual values data where representation as colors

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x =  flights.pivot_table(index='year', columns='month',values='passengers',aggfunc='sum')
```

```python
#heatmap
sns.heatmap(x)
```

```python
#linewidth
sns.heatmap(x,linewidths='2')
```

```python
#remove bar
sns.heatmap(x,linewidths='2', cbar=False)
```

```python
#set annote true
sns.heatmap(x,linewidths='2', cbar=False,annot=True)
```

```python
##set int type
sns.heatmap(x,linewidths='2', cbar=False,annot=True, fmt='d')
```

```python
#colors in heatmap
sns.heatmap(x,linewidths='2', cbar=False,annot=True, fmt='d',cmap='summer')
```

#### Clustermap

```python
sns.clustermap(x)
```

```python
#finding distnace with different formula
sns.clustermap(x,metric='correlation')
```

```python
#normalize z score
sns.clustermap(x,metric='correlation',z_score=0)
```

```python
#annotations = True
sns.clustermap(x,metric='correlation',z_score=0,annot = True)
```

```python
#only row wise clustering
sns.clustermap(x,metric='correlation',z_score=0,annot = True,col_cluster=False)
```

```python
#only column wise clustering
sns.clustermap(x,metric='correlation',z_score=0,annot = True, row_cluster=False)
```

#### Distplots
it is a combination of 3 graphs
1. histogram
2. kernal density estimation(PDF)
3. Rugplot

```python
titanic= pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(),inplace=True)
```

```python
#histogram
sns.displot(titanic['Age'],bins=15)
```

```python
#distplot with kde
sns.displot(titanic['Age'],kde=True,bins=10)
```

```python
#rugplot
sns.displot(titanic['Age'],rug=True)
```

```python
sns.displot(titanic[titanic['Survived']==1]['Age'],kde=True)
sns.displot(titanic[titanic['Survived']==0]['Age'],kde=True)
```
